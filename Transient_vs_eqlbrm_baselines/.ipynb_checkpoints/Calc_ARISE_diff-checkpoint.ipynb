## Transient vs QE baslines for estimating the impact of SAI, under the ARISE scenario in UKESM

Author: Alistair Duffey

alistair.duffey.21@ucl.ac.uk

#### Inputs:
1) a set of net cdf files containing maps of mean-state of a variable under some time-slice for each ensemble member and scenario, where in each case, the time slice is selected to match gmst during a corresponding window in another scenario. The two slices will have different rate of gmst change (here, R). 
2) a corresponding list of the values of R under each scenario and ensemble member. 

#### Outputs:
1) An estimate of the pattern effect of anomaly in a given variable due to the rate of change of gmst, relative to the value of that variable at Quasi-equilibrium.
2) Map of estimated changes under SAI (ARISE) at 1.5C with respect to a hypothetical QE baseline, contrasted with the same map as measured against the modelled, transient, 1.5C baseline.

#### Limitations:
1) no way to account for the dependence of anomaly on the absolute value of warming at the point of measuring the transient effect. Here we assume simple linear proportionality to the warming wrt pre-ind. 

```python

import os
import glob
import iris
import pandas as pd
import numpy as np
import logging
import esmvalcore.preprocessor
import xarray as xr
import xesmf as xe
from tqdm import tqdm
from xmip.preprocessing import rename_cmip6
import matplotlib.pyplot as plt
import dask
from nc_processing import calc_spatial_mean
from tqdm import tqdm
import scipy.stats as stats

import cartopy.feature as cfeature
import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
from cartopy.util import add_cyclic_point
import seaborn as sns

def read_in(dir, ocean = False):
    files = []
    for x in os.listdir(dir):
        files.append(dir + x)
    with dask.config.set(**{'array.slicing.split_large_chunks': True}):
        ds = rename_cmip6(xr.open_mfdataset(files, use_cftime=True, engine='netcdf4'))
    return ds

def read_in_ens_mean(dirs, ocean = False):
    """ returns (1) the ensemble mean dataset, and (2) the number of ensemble members """
    
    files = []
    for dir in dirs:
        for x in os.listdir(dir):
            if '.nc' in x:
                files.append(dir + x)
    with dask.config.set(**{'array.slicing.split_large_chunks': True}):
        ds = rename_cmip6(xr.open_mfdataset(files, use_cftime=True, concat_dim='ensemble',combine='nested'))
        n_ens = len(ds.ensemble) 
        ds = ds.mean(dim='ensemble')
        ds['number_ens_mems_meaned'] = n_ens
    return ds, n_ens

def get_gmst(ds):
    return calc_spatial_mean(ds.tas.mean(dim="time"), lon_name="x", lat_name="y").values
```

```python
### set options
var = 'tas'
base_scen = 'ssp126'
make_abs_temp_correction = False # if True we assume the that d(var)/d(R) is proportional to gmst above pre-ind


### constants
var_unit_conversions = {'pr':86400, #kgm-2s-1 to mm/day
                        'tas':1}
units = {'pr':'mm/day', 'tas':'°C'}[var]
```

```python
### read in data with pre-saved values of warming through the window
windows_df = pd.read_csv('int_outs_ssps/windows_df.csv')
windows_df = windows_df[windows_df['Model']=='UKESM1-0-LL']
```

```python
### make datasets with all UKESM maps, with variable for experiment, ensemble member and warming rate

def make_ds(path, drop=False):
    ds_list = []
    paths = glob.glob(path)
    if drop:
        paths = [x for x in paths if base_scen not in x ]
    for file_path in paths:
        ds = rename_cmip6(xr.open_dataset(file_path))        
        exp = file_path.split('_')[3]
        ds['Experiment'] = exp
        ds['Exp_Ens'] = exp+str(ds.Ensemble_member.values)
        R = (windows_df[windows_df['Scenario']==exp][windows_df['Ensemble_member']==str(ds.Ensemble_member.values)]['T_gradient']).values[0]
        ds['R'] = R
        ds_list.append(ds)
    DS = xr.concat(ds_list, dim='Exp_Ens')
    return DS

DS_all = make_ds('int_outs_ssps/UKESM1-0-LL_*_{v}/UKESM1-0-LL_*_{v}.nc'.format(v=var))
DS_base = make_ds('int_outs_ssps/UKESM1-0-LL_{s}_{v}/UKESM1-0-LL_*_{v}.nc'.format(s=base_scen, v=var)).mean(dim='Exp_Ens')
DS_trans = make_ds('int_outs_ssps/UKESM1-0-LL_*_{v}/UKESM1-0-LL_*_{v}.nc'.format(v=var), drop=True)

DS_trans_anom = DS_trans.copy()
DS_trans_anom[var] = DS_trans[var]-DS_base[var]
```

```python
len(DS_trans_anom.Exp_Ens)
```

```python

def calculate_correlation(datasets):
    correlations = pd.DataFrame(index=range(len(datasets)), columns=range(len(datasets)))
    
    for i in range(len(datasets.Exp_Ens)):
        for j in range(len(datasets.Exp_Ens)):
            corr_value = np.round(xr.corr(datasets[var].isel(Exp_Ens=i), 
                                 datasets[var].isel(Exp_Ens=j), 
                                 dim=['x', 'y']).compute().values, 5)
            correlations.loc[i, j] = corr_value
            correlations.loc[j, i] = corr_value
    
    return correlations
correlation_table = calculate_correlation(DS_trans_anom)

""" Spatial correlations between all maps of anomaly
    relative to SSP126 at the same temperature. High 
    correlations indicate that my defining a map of 
    d(var)/d(R) is reasonable """

correlation_table
```

```python
### get linear regression on R, take the m terms from y = mx+c
# N.B we might want to replace the below with a linear regression forced through zero
x = DS_trans_anom.swap_dims({'Exp_Ens': 'R'})
m = x[var].polyfit('R', deg=1).sel(degree=1)
to_plot = m.polyfit_coefficients*var_unit_conversions[var]/10
to_plot.plot(robust=True)
plt.title('d({v})/d(R), [{u}]/[°C per decade]'.format(v=var, u=units))
```

apply the map above of d(var)/d(R) to the ARISE simulations, to estimate how the changes relative to baseline would be different if using a QE baseline at 1.5C

```python
ens_mems = ['r1i1p1f2', 'r2i1p1f2', 'r3i1p1f2', 'r4i1p1f2', 'r8i1p1f2']
ens_mems = ['r1i1p1f2', 'r2i1p1f2', 'r3i1p1f2', 'r8i1p1f2']
```

```python
### make ARISE ensemble mean ds
ds_list = []
paths = glob.glob('/badc/deposited2022/arise/data/ARISE/MOHC/UKESM1-0-LL/arise-sai-1p5/*/Amon/{}/gn/*/'.format(var))

for path in paths:
    #files = os.listdir(path)
    ds = rename_cmip6(xr.open_mfdataset(path+'*.nc'))
    ds_list.append(ds)

DS_ARISE = xr.concat(ds_list, dim='Ensemble_member').mean(dim='Ensemble_member')
SAI = DS_ARISE.sel(time=slice('2050', '2070')).mean(dim='time')
```

```python
### make ssp245 ensemble mean ds (for appropriate ensemble members)
years = ['2015', '2033']

def get_ssp245_ds(variable):
    ds_list = []
    for es in ens_mems:
        path = '/badc/cmip6/data/CMIP6/ScenarioMIP/MOHC/UKESM1-0-LL/ssp245/{e}/Amon/{v}/gn/latest/'.format(e=es, v=variable)
        #files = os.listdir(path)
        ds = rename_cmip6(xr.open_mfdataset(path+'*.nc'))
        ds_list.append(ds)
    
    DS = xr.concat(ds_list, dim='Ensemble_member').mean(dim='Ensemble_member')
    return DS
    
DS_ssp245 = get_ssp245_ds(var)
baseline_tr = DS_ssp245.sel(time=slice(years[0], years[1])).mean(dim='time') # years are taken to match Henry et al. 2023

DS_tas_ssp245 = get_ssp245_ds('tas').sel(time=slice(years[0], years[1]))
slope_gmst_ssp245 = calc_spatial_mean(DS_tas_ssp245.tas, lon_name="x", lat_name="y").groupby('time.year').mean('time')

slope = stats.linregress(slope_gmst_ssp245.year, slope_gmst_ssp245).slope
```

```python
### define an estimated QE baseline based on the linear regression above

### find the overshoot peak temp
peak_temp = 2.8 ## roughly - TO FIX
if make_abs_temp_correction:
    abs_temp_correction = 1.5/peak_temp
else: 
    abs_temp_correction = 1

baseline_qe = (baseline_tr[var] - m.polyfit_coefficients*abs_temp_correction*slope).to_dataset(name=var)
```

```python
(baseline_tr[var]*var_unit_conversions[var]).plot()
```

```python
### plot the difference in baselines
TR_to_plot = (baseline_tr[var]*var_unit_conversions[var]) 
QE_to_plot = (baseline_qe[var]*var_unit_conversions[var])

diff_to_plot = 100*(TR_to_plot - QE_to_plot)/TR_to_plot

if var == 'tas':
    diff_to_plot = (TR_to_plot - QE_to_plot)


cmaps = {'tas':'RdBu_r', 
         'pr':'BrBG'}

# Define the contour levels to use in plt.contourf
if var == 'tas':
    clevs=np.arange(-1,1.2,0.2) 
elif var == 'pr':
    clevs=np.arange(-10,11,1) 

nrows, ncols = 1, 3
if var=='tas':
    nrows, ncols = 1, 1
fig, axs = plt.subplots(nrows=nrows,ncols=ncols,
                        subplot_kw={'projection': ccrs.PlateCarree()},
                        figsize=(11,4))
if nrows >1 and ncols > 1:
    axs=axs.flatten()
datasets = [TR_to_plot, QE_to_plot, diff_to_plot]

if var == 'pr':
    titles = ['Transient baseline (mm/day)', 'Estimated QE baseline (mm/day)', 'Difference (%)']
else:
    titles = ['Difference (°C)']
    datasets = [diff_to_plot]

i=0
for data in datasets:
    data,lons=add_cyclic_point(data,coord=data['x'])
    if var =='tas':
        ax = axs
    else:
        ax = axs[i]
    cs=ax.contourf(lons,ds['y'],data,
                       clevs,
                      transform = ccrs.PlateCarree(),
                      cmap=cmaps[var],
                       extend='both')
    ax.set_title(titles[i])
    ax.coastlines()
    i=i+1

if var=='tas':
    cbar = fig.colorbar(cs, label='°C')
else:
    fig.subplots_adjust(bottom=0.2, top=0.9, left=0.1, right=0.9,
                    wspace=0.02, hspace=0.02)
    cbar_ax = fig.add_axes([0.25, 0.2, 0.5, 0.03])
    cbar=fig.colorbar(cs, cax=cbar_ax,orientation='horizontal', label='{v}'.format(v=var))

plt.suptitle('Baseline mean-states, {}'.format(var))
#plt.tight_layout()
plt.savefig('Figures/baselines_{}.jpg'.format(var), dpi=450)
```

```python
### make a nice plot
TR_to_plot = ((SAI-baseline_tr)[var]*var_unit_conversions[var])
QE_to_plot = ((SAI-baseline_qe)[var]*var_unit_conversions[var])
diff_to_plot = TR_to_plot - QE_to_plot

 

cmaps = {'tas':'RdBu_r', 
         'pr':'BrBG'}

# Define the contour levels to use in plt.contourf
if var == 'tas':
    clevs=np.arange(-2,2.2,0.2) 
elif var == 'pr':
    clevs=np.arange(-0.8,0.9,0.2) 

nrows, ncols = 1, 3

# Define the figure and each axis for the 3 rows and 3 columns
fig, axs = plt.subplots(nrows=nrows,ncols=ncols,
                        subplot_kw={'projection': ccrs.PlateCarree()},
                        figsize=(11,4))

# axs is a 2 dimensional array of `GeoAxes`.  We will flatten it into a 1-D array
if nrows >1 and ncols > 1:
    axs=axs.flatten()

datasets = [TR_to_plot, QE_to_plot, diff_to_plot]
titles = ['Transient baseline', 'Estimated QE baseline', 'Difference']

i=0
for data in datasets:
    
    
    # Add the cyclic point
    data,lons=add_cyclic_point(data,coord=data['x'])
    
    # Contour plot
    cs=axs[i].contourf(lons,ds['y'],data,
                       clevs,
                      transform = ccrs.PlateCarree(),
                      cmap=cmaps[var],
                       extend='both')
    
    # Title each subplot with the name of the model
    axs[i].set_title(titles[i])
    
    # Draw the coastines for each subplot
    axs[i].coastlines()
    i=i+1


# Adjust the location of the subplots on the page to make room for the colorbar
fig.subplots_adjust(bottom=0.2, top=0.9, left=0.1, right=0.9,
                    wspace=0.02, hspace=0.02)

# Add a colorbar axis at the bottom of the graph
cbar_ax = fig.add_axes([0.25, 0.2, 0.5, 0.03])

# Draw the colorbar
cbar=fig.colorbar(cs, cax=cbar_ax,orientation='horizontal', label='{v} anom. {u}'.format(v=var, u=units))

# Add a big title at the top
plt.suptitle('ARISE 2050-2070 - SSP245 2015-2033, {}'.format(var))
#plt.tight_layout()
plt.savefig('Figures/ARISE_estimated_TR_vs_QE_anoms_{}.jpg'.format(var), dpi=450)
```

Calculate the spatial correlation between 
1) Tr-QE baseline, and 
2) residual changes measured against Tr baseline.

This spatial correlation should describe how much of the pattern of residual changes under SAI as measured against the Tr baseline, is actually explained by relaxation towards QE baseline

```python
da_1 = baseline_tr[var] - baseline_qe[var]
da_2 = SAI[var]-baseline_tr[var]

r = xr.corr(da_1, da_2, dim=['x', 'y']).compute()
print(r)
```
